```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('all_result.csv')

fig, ax = plt.subplots()
x= df['epoch'][:150]
ax.plot(x, df['1_hiddenlayer_ANN'][:150], label='1_hiddenlayer_ANN')
ax.plot(x, df['5_hidden_layer_ANN'][:150], label='5_hidden_layer_ANN')
ax.plot(x, df['lenet'][:150], label='lenet')
ax.plot(x, df['custom_CNN'][:150], label='custom_CNN')

ax.set_xlabel('Epoch')
ax.set_ylabel('Validate Accuracy')
ax.legend()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('loss.csv')

fig, ax = plt.subplots()

ax.plot(df['epoch'], df['1_hiddenlayer_ANN'], label='1_hiddenlayer_ANN')
ax.plot(df['epoch'], df['5_hidden_layer_ANN'], label='5_hidden_layer_ANN')
ax.plot(df['epoch'], df['lenet'], label='lenet')
ax.plot(df['epoch'], df['custom_CNN'], label='custom_CNN')

ax.legend()
ax.set_xlabel('Epoch')
ax.set_ylabel('Validate Loss')

# Tm ch s ca gi tr max
idx1 = df['1_hiddenlayer_ANN'].idxmin()
idx2 = df['5_hidden_layer_ANN'].idxmin()
idx3 = df['lenet'].idxmin()
idx4 = df['custom_CNN'].idxmin()

# Ly gi tr ti ch s 
max1 = df.loc[idx1, '1_hiddenlayer_ANN']
max2 = df.loc[idx2, '5_hidden_layer_ANN']
max3 = df.loc[idx3, 'lenet']
max4 = df.loc[idx4, 'custom_CNN']

# V annotate ti v tr epoch tng ng
ax.annotate(f"{max1:.3f}", xy=(idx1, max1), xytext=(idx1-10, max1+0.07))
ax.annotate(f"{max2:.3f}", xy=(idx2, max2), xytext=(idx2-10, max2+0.05))
ax.annotate(f"{max3:.3f}", xy=(idx3, max3), xytext=(idx3-10, max3+0.03))
ax.annotate(f"{max4:.3f}", xy=(idx4, max4), xytext=(idx4-8, max4-0.06))
ax.set_xlim(0) # Thm dng ny
ax.set_ylim(0) # Thm dng ny
plt.show()
```
